```python


import requests
from bs4 import BeautifulSoup

link_temp = 'https://www.allsides.com/story/middle-east-top-cia-official-posted-pro-palestine-image-her-facebook-profile'

def scraping_temp(link_):
    left_article_title = []
    left_article_publisher = []
    left_article_link = []


    center_article_title = []
    center_article_publisher = []
    center_article_link = []

    right_article_title = []
    right_article_publisher = []
    right_article_link = []
    
    response = requests.get(link_)
    soup = BeautifulSoup(response.text, 'html.parser')
    for elem in soup.find_all('div', attrs={'class': 'news-item left'}):
        link = elem.find('a')['href']
        article_text = elem.find('a').text
        dis = elem.find('span').text
        left_article_title.append(article_text)
        left_article_publisher.append(dis)
        left_article_link.append(link)

    for elem in soup.find_all('div', attrs={'class': 'news-item center'}):
        link = elem.find('a')['href']
        article_text = elem.find('a').text
        dis = elem.find('span').text
        center_article_title.append(article_text)
        center_article_publisher.append(dis)
        center_article_link.append(link)

    for elem in soup.find_all('div', attrs={'class': 'news-item right'}):
        link = elem.find('a')['href']
        article_text = elem.find('a').text
        dis = elem.find('span').text
        right_article_title.append(article_text)
        right_article_publisher.append(dis)
        right_article_link.append(link)


    left_df = pd.DataFrame(list(zip(left_article_title, left_article_publisher, left_article_link)))
    left_df['classifier'] = 'left'
    center_df = pd.DataFrame(list(zip(center_article_title, center_article_publisher, center_article_link)))
    center_df['classifier'] = 'center'
    right_df = pd.DataFrame(list(zip(right_article_title, right_article_publisher, right_article_link)))
    right_df['classifier'] = 'right'

    return pd.concat([left_df, center_df, right_df]).reset_index(drop=True)
```
